```python
import numpy as np
import pandas as pd
import fitter as ft
import datetime
```

```python
start = datetime.datetime.now()
beer = pd.read_csv('beer_reviews_updated.csv')
beer
```

```python
def fitDistribution(df,colname):
    var = df[colname].values
    dist = ft.Fitter(var)
    dist.fit()
    return dist
```

```python
#review_overall
ro = fitDistribution(beer,'review_overall')
ro.summary()
```

```python
#review_aroma
rar = fitDistribution(beer,'review_aroma')
rar.summary()
```

```python
#review_appearance
rapp = fitDistribution(beer,'review_appearance')
rapp.summary()
```

```python
#review_palate
rapa = fitDistribution(beer,'review_palate')
rapa.summary()
```

```python
#review_taste
rt = fitDistribution(beer,'review_taste')
rt.summary()
```

```python
#beer_abv
ba = fitDistribution(beer,'beer_abv')
ba.summary()
```

```python
end = datetime.datetime.now()
print("The program lasted: {} to execute".format(end-start))
```
